```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# you can use tkinter to allow the user to open a file browser and select an image
# check if you have tkinter install by typing in the command line "python -m tkinter"
# if a test tk GUI appears then you have it. Otherwise "pip install tkinter"

#import tkinter as tk 
#from tkinter import filedialog
#root = tk.Tk() 
#root.withdraw() 
#file_path = filedialog.askopenfilename() 
#img = cv2.imread(file_path) 
 

# use window thread to allow closing windows
cv2.startWindowThread()

# provide the path to your image file
img_bgr=cv2.imread('../../images/peppers.png')

cv2.imshow('input image', img_bgr)

print('press a key to plot histograms')
cv2.waitKey()


# convert from BGR to RGB because cv2 reads in the BGR order
img=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)

# obtain the color histograms
red_hist = cv2.calcHist([img], [0], None, [256], [0, 255])
green_hist = cv2.calcHist([img], [1], None, [256], [0, 255])
blue_hist = cv2.calcHist([img], [2], None, [256], [0, 255])


# plot the histograms
plt.subplot(4, 1, 1)
plt.imshow(img)
plt.title('image')
plt.xticks([])
plt.yticks([])

plt.subplot(4, 1, 2)
plt.plot(red_hist, color='r')
plt.xlim([0, 255])
plt.title('red histogram')

plt.subplot(4, 1, 3)
plt.plot(green_hist, color='g')
plt.xlim([0, 255])
plt.title('green histogram')

plt.subplot(4, 1, 4)
plt.plot(blue_hist, color='b')
plt.xlim([0, 255])
plt.title('blue histogram')

plt.tight_layout()
plt.show()

print('press a key to close image')
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)

# if you would like to see the image as read by cv2 
print('press a key to show original image inline (BGR format)')
cv2.waitKey()
plt.imshow(img_bgr)
```
