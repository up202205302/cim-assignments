```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#img = Image.open('../../images/peppers.png')
#M = np.asarray(img)

M=cv2.imread('../../images/peppers.png')

cv2.imshow('input image', M)

# if you would like to see the image as read by cv2 
print('press a key to show original image inline (BGR format)')
cv2.waitKey()
plt.imshow(M)
plt.title("Original image as read by cv2 (BGR)")


plt.figure(figsize=(12, 6))
plt.subplot(131)

plt.imshow(M[:, :, 0], cmap='Reds', vmin=0, vmax=255)
plt.title("Red Channel")
plt.subplot(132)

plt.imshow(M[:, :, 1], cmap='Greens', vmin=0, vmax=255)
plt.title("Green Channel")
plt.subplot(133)

plt.imshow(M[:, :, 2], cmap='Blues', vmin=0, vmax=255)
plt.title("Blue Channel")

plt.show()

# now by using directly each color channel

redChannel=M[:,:,2]
greenChannel=M[:,:,1]
blueChannel=M[:,:,0]

plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(redChannel)
plt.title("Red Channel")

plt.subplot(132)
plt.imshow(greenChannel)
plt.title("Green Channel")

plt.subplot(133)
plt.imshow(blueChannel)
plt.title("Blue Channel")

plt.show()

print('press a key to close image')
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
