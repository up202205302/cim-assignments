```python
# use tkinter to allow the user to open a file browser and select an image
# check if you have tkinter install by typing in the command line "python -m tkinter"
# if a test tk GUI appears then you have it. Otherwise "pip install tkinter"

import tkinter as tk 
from tkinter import filedialog
import cv2 
import matplotlib.pyplot as plt 

root = tk.Tk() 
root.withdraw() 
file_path = filedialog.askopenfilename() 

img = cv2.imread(file_path) 
  
# shows the image inline.
# Note that with plt image will be presented as BGR
plt.imshow(img)

# shows the image in a separate window for 2s and then closes it.
# Note that with cv2 image will be presented as RGB
cv2.imshow('RGB Image',img)
cv2.waitKey(2000) # wait 2000 ms 
cv2.destroyWindow('RGB Image')
cv2.waitKey(1)

img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

# or wait for the user to press a key before closing windows
cv2.imshow('Gray Image',img_gray)
cv2.waitKey() 
cv2.destroyWindow('Gray Image')
cv2.waitKey(1)

img_bgr=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)

# wait for the previous window to close and display a new converted image
# wait for the user to press a key before closing opened windows
cv2.imshow('BGR Image',img_bgr)
cv2.waitKey() 
cv2.destroyWindow('BGR Image')
cv2.waitKey(1)
```
