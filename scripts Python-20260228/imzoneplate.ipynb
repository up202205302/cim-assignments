```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Set the size of the image
size = 2048

# Generate the x and y coordinates of the image
x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))

# Generate the zoneplate image
r = np.sqrt(x**2 + y**2)
z = np.cos(20 * np.pi * r**2)

# Display the image
plt.imshow(z, cmap='gray')
plt.show()

# reshape to 2D, save as PIL image and display
mat = np.reshape(z,(size,size))
img = Image.fromarray(np.uint8(z * 255) , 'L')
img.show()
```
