```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski
from skimage import io, color

# Load the original image
img = Image.open("images/praia.bmp")
img.show()

img1=ski.io.imread("images/praia.bmp")
plt.imshow(img1)
plt.title('input image')
plt.show()

# Convert the image to a numpy array
img_arr = np.array(img)

# Separate the RGB components
r = img_arr[:,:,0]
g = img_arr[:,:,1]
b = img_arr[:,:,2]

# Display each component
#Image.fromarray(r).show()
#Image.fromarray(g).show()
#Image.fromarray(b).show()

plt.imshow(r, cmap='gray')
plt.title('r image')
plt.show()

plt.imshow(g, cmap='gray')
plt.title('g image')
plt.show()

plt.imshow(b, cmap='gray')
plt.title('b image')
plt.show()

# Convert the image to the HSV color space
hsv_img = img.convert('HSV')
hsv_img.show()
img.save("praia_hsv.bmp")

# Convert the HSV image to a numpy array
hsv_arr = np.array(hsv_img)

# Separate the HSV components
h = hsv_arr[:,:,0]
s = hsv_arr[:,:,1]
v = hsv_arr[:,:,2]

# Display each component
#Image.fromarray(h).show()
#Image.fromarray(s).show()
#Image.fromarray(v).show()
plt.imshow(h, cmap='gray')
plt.title('h image')
plt.show()

plt.imshow(s, cmap='gray')
plt.title('s image')
plt.show()

plt.imshow(v, cmap='gray')
plt.title('v image')
plt.show()
```
